```python
bins = [0,5,10,20]
labels = ['Low', 'Medium', 'High']
df['Experience Category'] = pd.cut(df['experience'], bins = bins, labels = labels, include_lowest=True)

df_encoded = pd.get_dummies(df.drop(columns = ['experience']))

df_encoded.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

min_support = 0.1
frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)

min_confidence = 0.7
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold=min_confidence)

print(frequent_itemsets)
print(rules)
```

```python
sorted_rules_support = rules.sort_values(by = 'support', ascending=False)

sorted_rules_confidence = rules.sort_values(by = 'confidence', ascending = False)

sorted_rules_lift = rules.sort_values(by = 'lift', ascending = False)

```
